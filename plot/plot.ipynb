```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
metric = 'EM'
```

```python
df_b = pd.read_csv("run_train_baseline-02-tag-dev_%s.csv"%metric)
df_s = pd.read_csv("run_train_BiDAF_syll-03-tag-dev_%s.csv"%metric)
df_c = pd.read_csv("run_train_BiDAF_char-04-tag-dev_%s.csv"%metric)
df_sc = pd.read_csv("run_train_BiDAF_both-01-tag-dev_%s.csv"%metric) 
```

```python
df_b.head()
```

```python
x_scale = 1000
plt.plot(df_b.Step / x_scale, df_b.Value)
plt.plot(df_s.Step / x_scale, df_s.Value)
plt.plot(df_c.Step / x_scale, df_c.Value)
plt.plot(df_sc.Step /x_scale, df_sc.Value)

plt.ylabel(metric)
plt.xlabel('k-iterations')
plt.legend(["baseline", "syllable", "character", "syllable + character"])

plt.savefig("%s.png"%metric)
```
